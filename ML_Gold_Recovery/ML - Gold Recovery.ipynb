```python
# Универсальная функция для исследования проблем датасета 
def explore_dataset(dataset):
    from IPython.display import display
    
    print('Первые пять строк датасета:')
    display(dataset.head(5))
    
    
    print('')
    print(f'Датасет имеет размер {dataset.shape[0]} строк и {dataset.shape[1]} столбцов.')
    missing_columns = dataset.isna().any()[lambda x: x]
    missing_values = dataset.isnull().sum()/len(dataset)
    missing_values = missing_values[missing_columns.keys()].sort_values(ascending = False)
    if len(missing_columns)>0:
        print(f'Из них {len(missing_columns)} столбцов имеют пропуски в данных.')
        print(f'Это столбцы: ', ', '.join(dataset.isna().any()[lambda x: x].keys()))
        print('')    
        bars = missing_values.plot(kind = 'barh', figsize=(10,round(len(missing_columns)/3)))
        plt.title('Процент пропущенных значений по столбцам')
        for i, v in enumerate(missing_values):
            plt.gca().text(v + 0.005*missing_values.max(), i - 0.1, '{:.2%}'.format(v), color='black')
        plt.show()
        plt.close()
    else:
        print('Пропусков в данных нет')
        
        
    if len(dataset[dataset.duplicated()])>0:
        print(f'В датасете содержится {len(dataset[dataset.duplicated()])} дубликатов')
    else:
        print('В данных не найдено полных дубликатов.')
        
        
    print('')
    print('Столбцы имеют следующие типы данных:')    
    print(dataset.dtypes)
    
    
    print('')
    print('Общая статистика по количественным переменным:')    
    display(dataset.describe())
    print('')
    print('Общая статистика по категориальным переменным:')    
    display(dataset.describe(include=["object", "bool"]))
    
    
    
    return
```

### Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery. Найдите MAE между вашими расчётами и значением признака. Опишите выводы.

Напишем функцию для расчета обогащения:

```python
def recovery(c, f, t):
    return c * (f - t)/ (f * (c - t)) * 100
```

```python
recovery_df = data_train.apply(lambda x: recovery(x['rougher.output.concentrate_au'], 
                                                  x['rougher.input.feed_au'],
                                                  x['rougher.output.tail_au']), 
                               axis=1)
```

```python
recovery_df.hist()
```

```python
recovery_mae = mean_absolute_error(data_train['rougher.output.recovery'],recovery_df)
print('MAE между расчётами и значением признака равно', recovery_mae)
```

Как видно из результата, ошибка между расчетным и табличным значением находится в пределах погрешности вычислений для чисел с плавающей точкой, а значит эффетивность обогащения в данных рассчитана верно. Стоит отметить, что в признаке эффектичности обогащения есть выбросы с нулевыми значениями, которые стоит проверить на этапе подготовки данных.

#### Обработаем аномалии в обучающем датасете

Посмотрим для начала распределение данных для целевых признаков:

```python
def plot_hist(data, col_column):
    '''
    Функция отрисовки гистограмм и ящика с усами для количественных переменных.
    На вход: исходная таблица и список количественных переменных.
    На выходе: графики
    '''
    rows = len(col_column)
    f, ax = plt.subplots(rows,2, figsize=(12, rows*3))
    f.tight_layout()
    plt.rcParams.update({'font.size': 18})
    
    for i, col in enumerate(col_column):         
        sns.histplot(data[col], kde=True, bins=50, ax = ax[i, 0])                    
        sns.boxplot(data[col], ax = ax[i, 1])

        ax[i, 0].set_xlabel(col)
        ax[i, 1].set_xlabel(col)
        ax[i, 0].set_ylabel('Количество')
    plt.suptitle("Гистограмма и ящик с усами для количественных данных", fontsize=22, y=1.01)
    plt.show()
```

```python
plot_hist(data_train, target_columns)
```

Как видим из графиков, эффетивность обогащения мжет принимать значения от 0 до 100, и в принципе ни одно из этих значений не будет аномальным. Однако, судя по гистограмме частота значений 0 для эффективности на черновом и финальном этапах, а также частота значения 100 на финальном этапе выбиваются из общего распределения признака. Посмотрим внимательней на причину таких значений:

```python
data_train.loc[data_train[target_columns[0]]==0]
```

```python
data_train.loc[data_train[target_columns[0]]==100]
```

Внимательно рассмотрев данные можно заметить, что ноль возникает, когда коцентрат продукта равен нулю, а значение 100 - когда концентрация золота в хвостах нулевая. При этом эти величины зануляются сразу для всех измеряемых веществ. Скорее всего такие ситуации вызваны какими-то сбоями в процессе обогащения, нежели чем реальной ситуацией. Поскольку такие велечины не зависят от исходной концентрации сырья.  

Можно предположить, что такие сбои возможны на любом этапе очистки, проверим расспределение данных во всех столцах описывающих продукт после этапа обработки ('output' в названии столбца):

```python
plot_hist(data_train, [x for x in data_train.columns if 'output' in x])
```

Действительно, аномалии вокруг значения ноль присутствуют практическе в каждом признаке по концентрации веществ. Такие сбои в процессе обогащения могут влиять на результат модели, поскольку оценка эффективности обогащения не будет достоверной. Удалим все строки с нулями в данных признаках. Кроме того, как было выявлено позднее на этапе первичной очистки вместо нулей часто встречается значение 0.01, на гистограмме выглядит как пик около нуля, но на самом деле все значения в строках для концентрации веществ стоят - 0.01. Возможно техническая особенность, удалим такие данные тоже.

```python
print('Размер обучающей выборки до удаления аномалий:', data_train.shape[0], 'строк' )
shape_old = data_train.shape[0]

for x in data_train.columns:
    if 'output' in x:
        data_train = data_train.drop(data_train[data_train[x]==0].index)
        
for x in data_train.columns:
    if 'primary_cleaner.output' in x:
        data_train = data_train.drop(data_train[data_train[x]<0.011].index)
        
print('Размер обучающей выборки после удаления аномалий:', data_train.shape[0], 'строк' )    
print('Удалено ', round((shape_old - data_train.shape[0]) / shape_old * 100, 2), '% данных' ) 
```

```python
plot_hist(data_train, target_columns)
```

Таким образом мы избавились от аномальных нулевых значений концентраций, а так же от нулевых и равных 100 значений эффективности. Теперь расспределение целевых признаков выглядит нормальным.

## Модель

### Напишите функцию для вычисления итоговой sMAPE.


```python
def smape(A, F):
    # функция для вычисления sMAPE
    # A - actual value
    # F - forecast value
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))

def final_smape(smape_rough, smape_final):
    # функция для расчета итогового sMAPE
    return 0.25 * smape_rough + 0.75 * smape_final
```

```python
# сделаем score для подбора наилучшей модели
smape_scorer = make_scorer(smape, greater_is_better=False)
```

Лучшие метрики при адекватном для учебного проекта времени обучения показали модели KNeighborsRegressor() и ExtraTreesRegressor(). Подберем наилучшую модель из них с помощью RandomizedSearchCV:

```python
def tune_model(features, target):
    
    clf1 = ExtraTreesRegressor( random_state=RANDOM_STATE)
    clf2 = KNeighborsRegressor()
    
    # Initiaze the hyperparameters for each dictionary
    param1 = {}
    param1['preprocessing'] =  [None, StandardScaler()]
    param1['classifier__max_depth'] =  list(np.arange(5, 100, step=5)) + [None]
    param1['classifier__criterion'] =  ['squared_error','absolute_error', 'friedman_mse', 'poisson']
    param1['classifier__max_features'] =  ['auto','sqrt','log2']
    param1['classifier__min_samples_leaf'] =   st.randint(1,4)
    param1['classifier__min_samples_split'] =   np.arange(2, 10, step=2)
    param1['classifier'] = [clf1]

    param2 = {}
    param2['preprocessing'] =  [None, StandardScaler()]
    param2['classifier__n_neighbors'] = np.arange(1, 20)
    param2['classifier__weights'] = ['uniform', 'distance']
    param2['classifier__algorithm'] = ['auto', 'ball_tree', 'kd_tree', 'brute']
    param2['classifier'] = [clf2]

   

    pipeline = Pipeline([('preprocessing', StandardScaler()),
                         ('classifier', clf1)])
    params = [param1, param2]
    
    cv = KFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE)
    grid =  RandomizedSearchCV(pipeline,
                               param_distributions=params,
                               scoring=smape_scorer,
                               n_iter=5,
                               cv=cv,
                               random_state=RANDOM_STATE,
                               refit=True,
                               verbose=1)

    grid.fit(features, target)

    print('Параметры наилучшей модели:')
    print("\n".join(f'{k} : {v}' for k,v in grid.best_params_.items()))
    print("Мера sMAPE для наилучшей модели:", round(-grid.best_score_,4))

    return grid.best_estimator_, -grid.best_score_
```

```python
# модель чернового концентрата
model_r, smape_r = tune_model(features_train_r, target_train_r)
```

```python
# модель финального концентрата
model_f, smape_f = tune_model(features_train_f, target_train_f)
```

```python
print("Мера итоговой sMAPE для наилучших моделей:",round(final_smape(smape_r, smape_f),2))
```

Наилучшей моделью была выбрана модель ExtraTreesRegressor для финального концентрата в сочетании с KNeighborsRegressor для чернового концентрата, они показали итоговую метрику sMAPE 4.75 на кросс-валидации.
