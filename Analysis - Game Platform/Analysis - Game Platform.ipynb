# Анализ популярности игровых платформ
## Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й.

**Описание данных**

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
# Универсальная функция для исследования проблем датасета 
def explore_dataset(dataset):
    from IPython.display import display
    
    print('Первые пять строк датасета:')
    display(dataset.head(5))
    
    
    print('')
    print(f'Датасет имеет размер {dataset.shape[0]} строк и {dataset.shape[1]} столбцов.')
    missing_columns = dataset.isna().any()[lambda x: x]
    missing_values = dataset.isnull().sum()/len(dataset)
    missing_values = missing_values[missing_columns.keys()].sort_values(ascending = False)
    if len(missing_columns)>0:
        print(f'Из них {len(dataset.isna().any()[lambda x: x])} столбцов имеют пропуски в данных.')
        print(f'Это столбцы: ', ', '.join(dataset.isna().any()[lambda x: x].keys()))
        print('')    
        bars = missing_values.plot(kind = 'barh', figsize=(10,6))
        plt.title('Процент пропущенных значений по столбцам')
        for i, v in enumerate(missing_values):
            plt.gca().text(v + 0.005, i - 0.1, '{:.2%}'.format(v), color='black')
        plt.show()
        plt.close()
    else:
        print('Пропусков в данных нет')
        
        
    if len(dataset[dataset.duplicated()])>0:
        print(f'В датасете содержится {len(dataset[dataset.duplicated()])} дубликатов')
    else:
        print('В данных не найдено полных дубликатов.')
        
        
    print('')
    print('Столбцы имеют следующие типы данных:')    
    print(dataset.dtypes)
    
    
    print('')
    print('Общая статистика по количественным переменным:')    
    display(dataset.describe())
    print('')
    print('Общая статистика по категориальным переменным:')    
    display(dataset.describe(include=["object", "bool"]))
    
    
    print('')
    print('Распределение и выбросы данных в количественных переменных:')    
    dataset.plot.box(subplots=True, 
                     layout=(1,len(dataset.select_dtypes(exclude=['object','bool']).columns)), 
                     vert='False',
                     figsize=(2.5*len(dataset.select_dtypes(exclude=['object','bool']).columns),8));
    plt.show()
    plt.close()
    return
```

Проверим может ли год выпуска одной игры на разных платформах быть разным:

```python
piv = data.pivot_table(index='name', 
                 values='year_of_release',
                 aggfunc=['min', 'max']
                )
piv.columns = ['min_year_of_release', 'max_year_of_release']
piv['difference'] = piv['max_year_of_release']-piv['min_year_of_release']
piv.loc[piv['difference']>0].sort_values('difference', ascending=False)
```

Действительно, в датасете 694 игры для которых год выпуска на разных платформах отличается на величины от 1 года до 34 лет. А значит заполнить пропуски в этом столбце на основе имеющихся данных для других платформ будет неправильно. Оставим пропуски как есть.

На рассмотрении остаются многочисленные пропуски в столбцах сritic_score, user_score, rating.

```python
print('В столбце critic score',len(data[data['critic_score'].isna()]), 'пропусков')
print('В столбце user_score',len(data[data['user_score'].isna()]), 'пропусков')
print('В столбце rating',len(data[data['rating'].isna()]), 'пропусков')
print('В трех столбцах одновременно',len(data.loc[(data['user_score'].isna())&(data['rating'].isna())&(data['critic_score'].isna())]), 'пропусков')
```

```python
count_nans = lambda x: len(x[x.isna()])
piv_critic_score = data.pivot_table(index='year_of_release',
                       values=['critic_score'],
                       dropna=False,
                       aggfunc=[len, count_nans]
                      )
piv_critic_score.columns = ['all_values', 'missing_values_critic_score']
```

```python
piv_user_score = data.pivot_table(index='year_of_release',
                       values=['user_score'],
                       dropna=False,
                       aggfunc=[ count_nans]
                      )
piv_user_score.columns = ['missing_values_user_score']
```

```python
piv_rating = data.pivot_table(index='year_of_release',
                       values=['rating'],
                       dropna=False,
                       aggfunc=[count_nans]
                      )
piv_rating.columns = ['missing_values_rating']
```

```python
piv_critic_score = piv_critic_score.join([piv_user_score ,piv_rating])
piv_critic_score.plot( figsize=(10,5), title='Количество игр и пропущенных значений в столбцах по годам');
```

Как мы видим, чаще всего пропуски присутствуют во всех трех столбцах одновременно. При этом  для игр примерно до 1995 года  отсутствует информация по рейтингу, оценкам критиков и пользователей. Согласно информации из интернета организация ESRB (англ. Entertainment Software Rating Board), начала присваивать рейтинг игр только с 1 сентября 1994.  В мае 1996 года появился интернет сайт посвященный обзорам видеоигр - GameSpot, а в 2001 году запустили самый крупный на данный момент сайт агрегирующий информацию о рейтингах и обзорах видеоигр - Metacritic. Такая историческая справка объясняет отсутствие данных в столбцах сritic_score, user_score, rating для игр выпущенных раньше 1994 года. 

Что касается более поздних годов, то резкое уеньшение пропусков в данных в 2000-2001 году связано с открытием сайта Metacritic, далее количество пропусков коррелирует с общим количеством выпускаемых по годам игр, а значит какой то процент от общего числа видеоигр все время остается без оценки пользователями и критиками. Скорее всего эти пропуски вызваны особенностями проставления оценок для игр, а не техническими ошибками в базе данных. Что касается рейтингов, то организация ESRB работает в Сша и Канаде, для Европы и Японии работают другие организации, рейтинг которых не представлен в данной базе данных. Этим объясняется большое количество игр без рейтинга.

Заполнение пропусков оценок критиков не имеет смысла, поскольку их наличие является особенностью этих параметров и любое искусственное заполнение исказит результаты исследования. Однако пропуски столбца rating лучше заполнить новым значением, к примеру, 'unrated', что будет означать что игре не присвоен рейтинг ESRB. 

```python
data['rating'].fillna('unrated', inplace=True)
```

```python
def pivot_table_by_region(dataframe, index, values):
    piv = actual_data.pivot_table(index=index,
                                  values=values,
                                  aggfunc= [np.sum, np.median, 'count']
                                 )
    piv.columns = ['sum','median','count']
    
    return piv
    

def plot_pie_from_pivot_table(piv, subplot, title, dict_colors, n=5):   
    
    # функция строит pie-chart для n наибольших по суммарному значению индексов датафрейма piv 
    
    # находим топ-n по сумме для значений index, остальные записываем в категорию Другие
    topn_index = piv.sort_values(by=['sum'], ascending=False).head(n).index
    piv['category'] = piv.index.to_series().apply(lambda x:
                                                  x if x in topn_index else 'Другие')
    # строим pie-chart
    grouped_piv = piv.groupby('category').sum().sort_values('sum')['sum']
    grouped_piv.plot(kind='pie', 
                     figsize=(15,10),
                     pctdistance = 0.5, 
                     labeldistance = 1.05,                     
                     autopct=lambda x: f'{x:.1f}%',
                     textprops={"size": 16},
                     radius = 2,
                     colors = [dict_colors[x] for x in grouped_piv.index], # находим соответствующий цвет из dict_colors
                                                                           # для всех index попавших в топ-N
                     explode=(grouped_piv.values == min(grouped_piv.values)) * 0.1,
                     ax = subplot
                    )
    subplot.axis('equal')
    subplot.set_ylabel('')
    subplot.set_title(title)
    subplot.legend(grouped_piv.index, loc="best", title="Platforms")

```

## Проверка гипотез


### Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевая гипотеза (Н_0) — предположение о том, что между  параметрами генеральных совокупностей нет различий, то есть эти различия носят не систематический, а случайный характер. Альтернативная гипотеза (Н_а) – предположение о том, что между параметрами генеральных совокупностей есть достоверные различия.

Таким образом, чтобы проверить гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC , сформулируем нулевую и альтернативную гипотезу следующим образом:

H_0: пользовательские рейтинги платформы Xbox One = пользовательские рейтинги платформы PC

H_a: пользовательские рейтинги платформы Xbox One ≠ пользовательские рейтинги платформы PC

В качестве метода для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам используем метод из библиотеки scipy: scipy.stats.ttest_ind().

Уровень значимости определим, как alpha = 0.05

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(actual_data[actual_data['platform'] =='XOne']['user_score'], 
                       actual_data[actual_data['platform'] =='PC']['user_score'],
                       equal_var=False,
                       nan_policy='omit')
# alpha = задайте значение уровня значимости
alpha = 0.05
# вывод значения p-value на экран 
print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue>alpha:
    print('Не получилось отвергнуть нулевую гипотезу')
else:
    print('Отвергаем нулевую гипотезу')
```

Нулевую гипотезу отергнуть не получилось, а значит различия между средними пользовательскими рейтингами платформ Xbox One и PC не значимы.

### Гипотеза: средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


Сформулируем нулевую и альтернативную гипотезы:


H_0: пользовательские рейтинги жанров Action = пользовательские рейтинги жанров Sports

H_a: пользовательские рейтинги жанров Action ≠ пользовательские рейтинги жанров Sports

Чтобы подтвердить наше предположение о различии рейтингов, неоходимо отвргнуть нулевую гипотезу.

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(actual_data[actual_data['genre'] =='Action']['user_score'], 
                       actual_data[actual_data['genre'] =='Sports']['user_score'],
                       equal_var=False,
                       nan_policy='omit')
# alpha = задайте значение уровня значимости
alpha = 0.05
# вывод значения p-value на экран 
print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue>alpha:
    print('Не получилось отвергнуть нулевую гипотезу')
else:
    print('Отвергаем нулевую гипотезу')
```

Мы отвергли нулевую гипотезу, а значит различия между средними пользовательскими рейтингами жанров Action и Sports значимые.
