# Выбор локации для скважины

Допустим, вы работаете в добывающей компании «ГлавРосГосНефть». Нужно решить, где бурить новую скважину.

Вам предоставлены пробы нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов. Постройте модель машинного обучения, которая поможет определить регион, где добыча принесёт наибольшую прибыль. Проанализируйте возможную прибыль и риски техникой *Bootstrap.*

Шаги для выбора локации:

- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.

```python
# Универсальная функция для исследования проблем датасета 
def explore_dataset(dataset):
    from IPython.display import display
    
    print('Первые пять строк датасета:')
    display(dataset.head(5))
    
    
    print('')
    print(f'Датасет имеет размер {dataset.shape[0]} строк и {dataset.shape[1]} столбцов.')
    missing_columns = dataset.isna().any()[lambda x: x]
    missing_values = dataset.isnull().sum()/len(dataset)
    missing_values = missing_values[missing_columns.keys()].sort_values(ascending = False)
    if len(missing_columns)>0:
        print(f'Из них {len(dataset.isna().any()[lambda x: x])} столбцов имеют пропуски в данных.')
        print(f'Это столбцы: ', ', '.join(dataset.isna().any()[lambda x: x].keys()))
        print('')    
        bars = missing_values.plot(kind = 'barh', figsize=(10,6))
        plt.title('Процент пропущенных значений по столбцам')
        for i, v in enumerate(missing_values):
            plt.gca().text(v + 0.005, i - 0.1, '{:.2%}'.format(v), color='black')
        plt.show()
        plt.close()
    else:
        print('Пропусков в данных нет')
        
        
    if len(dataset[dataset.duplicated()])>0:
        print(f'В датасете содержится {len(dataset[dataset.duplicated()])} дубликатов')
    else:
        print('В данных не найдено полных дубликатов.')
        
        
    print('')
    print('Столбцы имеют следующие типы данных:')    
    print(dataset.dtypes)
    
    
    print('')
    print('Общая статистика по количественным переменным:')    
    display(dataset.describe())
    print('')
    print('Общая статистика по категориальным переменным:')    
    display(dataset.describe(include=["object", "bool"]))
    
    
    
    return
```

```python
def plot_hist(data, col_column):
    '''
    Функция отрисовки гистограмм и ящика с усами для количественных переменных.
    На вход: исходная таблица и список количественных переменных.
    На выходе: графики
    '''
    rows = len(col_column)
    f, ax = plt.subplots(rows,2, figsize=(12, rows*3))
    f.tight_layout()
    plt.rcParams.update({'font.size': 18})
    
    for i, col in enumerate(col_column):         
        sns.histplot(data[col], kde=True, bins=24, ax = ax[i, 0])                    
        sns.boxplot(data[col], ax = ax[i, 1])

        ax[i, 0].set_xlabel(col)
        ax[i, 1].set_xlabel(col)
        ax[i, 0].set_ylabel('Количество')
    plt.suptitle("Гистограмма и ящик с усами для количественных данных", fontsize=22, y=1.01)
    plt.show()
```

## Обучение и проверка модели

Чтобы не повторять код, напишем функции для оценки качества модели и для процесса обучения модели, который включает в себя:

- разбиение выборки на обучающую и валидационную в соотношении 75%/25%
- обучение модели линейной регрессии на обучающей выборке
- расчет метрик качества модели на валидационной выборке
- расчет среднего запаса сырья по предсказанным и фактическим значениям
- сохранения фактического и предсказанного значений в датасет


```python
def metrics_model(fact, prediction):
    '''
    Функция выводит метрики качества для регрессионных моделей (точность модели).
    '''
       
    print('R2   = ', round(r2_score(fact, prediction),2))
    print('MAE  = ', round(mean_absolute_error(fact, prediction),2))
    print('MSE  = ', round(mean_squared_error(fact, prediction),2))
    print('RMSE = ', round(mean_squared_error(fact, prediction)**0.5,2))
```

```python
def fit_model_linreg(data, features_columns, target_column, test_ratio=0.25, random_state=12345):
    
    #Разбиваем выборку на обучающую и валидационную.
    features = data.drop([x for x in data.columns if x not in features_columns], axis=1)
    target = data[target_column]
    features_train, features_valid, target_train, target_valid =\
    train_test_split(features, target, test_size=test_ratio,
                     random_state=random_state)
    
    print('Размер обучающей выборки:', features_train.shape[0])
    print('Размер валидационной выборки:', features_valid.shape[0])
    
    #Обучаем модель линейной регрессии.
    model = LinearRegression()
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    
    print('Получили следующие метрики модели на валидационной выборке:')
    metrics_model(target_valid, predictions_valid)
    
    #Сохраняем предсказанные и правильные ответы.
    result = target_valid.copy()
    result['product_predicted'] = predictions_valid
    print('Средний запас предсказанного сырья:', round(predictions_valid.mean(),2),'тыс. баррелей')
    print('Средний запас фактического сырья:', round(target_valid['product'].mean(),2),'тыс. баррелей')
    
    return result
```

## Подготовка к расчёту прибыли

Условия задачи:

*При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
Бюджет на разработку скважин в регионе — 10 млрд рублей. При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей. После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.*

Запишем основные параметры задачи в коэффициенты:

```python
N_OIL_WELLS=200
BUDGET = 1e10
PRICE_PER_MBBL = 450000
LOSS_PROBABILITY = 0.025
```

Посчитаем минимальное среднее количество продукта в месторождениях региона, достаточное для разработки, для этого разделим бюджет разработки на количество скважин для разработки и стомость сырья за тыс. баррелей:

```python
min_product = BUDGET / (PRICE_PER_MBBL * N_OIL_WELLS)
print('Минимальное среднее количество продукта в месторождениях региона, достаточное для разработки:',
      round(min_product,2),'тыс. баррелей')
```

Полученное значение больше средних значений сырья по регионам и почти в два раза больше среднего значения во втором регионе. Посмотрим сколько скважин в каждом из регионов согласно предсказаниям имеет запас сырья больше минимального:

```python
print( 'Регион 1 :', result_data0.loc[result_data0['product_predicted'] >= min_product].count()[0],
      'скважин имеет запас сырья больше минимального')
print( 'Регион 2 :', result_data1.loc[result_data1['product_predicted'] >= min_product].count()[0],
      'скважин имеет запас сырья больше минимального')
print( 'Регион 3 :', result_data2.loc[result_data2['product_predicted'] >= min_product].count()[0],
      'скважин имеет запас сырья больше минимального')
```

В каждом из регионов скважин достаточно, а значит все регионы будем использовать для дальнейшего анализа.

Напишем функцию для расчёта прибыли по выбранным скважинам и предсказаниям модели:


```python
def profit_from_oil_wells(product, n=N_OIL_WELLS, price=PRICE_PER_MBBL, budget=BUDGET):
    wells_sorted = product['product_predicted'].sort_values(ascending=False)
    selected = product['product'][wells_sorted.index][:n]
    return  price * selected.sum() - budget
```

Полученное значение больше средних значений сырья по регионам и почти в два раза больше среднего значения во втором регионе, однако анализ показал, что каждом из регионов достаточно имеющих среднее значение запаса больше минимального. 

## Расчёт прибыли и рисков 

Напишем функцию для расчета прибыли и вероятности убытков с применением техники Bootstrap:

```python
def bootstrap(data, n_iter=1000, n_sample=500):
    state = np.random.RandomState(12345)
    
    values = []
    for i in range(n_iter):
        subset = data.sample(n_sample, replace=True, random_state=state)
        # для каждой случайной выборки считаем прибыль по 200-м скважинам с максимальным 
        # рассчитанным запасом за вычетом трат на разработку
        values.append(profit_from_oil_wells(subset))

    values = pd.Series(values)
    
    #находим 95% доверительный интервал
    lower = values.quantile(0.025)
    upper = values.quantile(0.975)
    
    # находим среднюю выручку
    mean = values.mean()
    
    # считаем вероятность значений прибыли меньше нуля
    loss_probability = ( values < 0 ).astype(int).mean()
    
    print("Средняя выручка:", round(mean/10e6,2), 'млн. руб.')
    print("95% доверительный интервал лежит между", round(lower/10e6,2),
          'и', round(upper/10e6,2),'млн. руб.')
    print("Вероятность убытков составляет ", round(loss_probability*100,2), '%')
    
```

Применим функцию для каждого региона:

```python
bootstrap(result_data0)
```

```python
bootstrap(result_data1)
```

```python
bootstrap(result_data2)
```

В первом и втором регионах риск убытков равен порядка 6% и больше заданного порогового значения в 2.5%. По всей видимости это связано с плохим качеством модели обученной на данных этих регионов. Модель линейной регрессии не смогла выявить линейные зависимости с запасом сырья в предложенных данных измерений и выдала ненадежные результаты предсказаний. Поэтому несмотря на то, что в первом и третьем регионов средний запас сырья наибольший, ошибки в предсказаниях слишком велики, чтобы надежно определится с разрабатываемыми скважинами.
