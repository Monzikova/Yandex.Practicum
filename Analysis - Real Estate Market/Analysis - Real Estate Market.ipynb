В квартирах элитного класса или старого фонда (сталинки) высота потолков значительно выше. Посмотрим на данные повнимательней:

#####  locality_name

```python
print(len(data[data['locality_name'].isna()]))
data[data['locality_name'].isna()]
```

В датасете 49 объявлений без указания населенного пункта. При этом во многих объявлениях указано расстояние от центра города, в нашем случае, это, предположительно, Санкт-Петербург.
Вероятная причина пропусков в этом столбце - техническая ошибка. Попробуем восстановить населенные пункты по имеющимся данным.

Для этого посмотрим как распределяются расстояния от центра города и от аэропорта по населенным пунктам:

```python
piv = data.pivot_table(index='locality_name', 
                 values=[ 'cityCenters_nearest', 'airports_nearest'],
                 aggfunc=['min', 'max']
                ).sort_values(by=('min', 'cityCenters_nearest'))
piv.columns = list(map("_".join, [x[::-1] for x in piv.columns]))
piv.sort_index(axis='columns', ascending=False)
```

Территория Санкт-Петербурга не представляет собой ровный круг, поэтому мы можем с некоторой уверенностью утверждать что квартиры с растоянием от центра города менее 17 км относятся к Санкт-Петербургу, с 17 до 29.5 км это может быть как Спб так и один из пригородов, а уже начиная с 30 км с большой долей вероятности это будет не Санкт-Петербург.

Учитывая вышесказанное, заменим пропуски в столбце locality_name по следующему правилу:
- если расстояние от центра меньше 17 км, то заменим на Санкт-Петербург
- если расстояние больше 17 км или неизвестно, то заменим на 'Неизвестно'.

```python
data.loc[data['cityCenters_nearest']<=17000., 'locality_name'].fillna('Санкт-Петербург', inplace=True)
data['locality_name'].fillna('Неизвестно', inplace=True)
```

```python
#проверим все ли пропуски заполнены
print(len(data[data['locality_name'].isna()]))
```

```python
# посмотрим еще раз на распределение величин после заполнения пропусков и преобразования типов данных
data.hist(figsize=(15, 20), grid=False, ec='black')
```

Построим диаграммы размаха для тех параметров, в которых наблюдаются выбросы согласно гистрограммам, а так же посчитаем границы статистически значимой выборки (без выбросов).

```python
fig, axes = plt.subplots(nrows = 2, ncols =3, figsize=(15,10) )
columns = ['total_images', 'floors_total', 'last_price', 'total_area', 'rooms', 'ceiling_height']
for i in range(6):
    data.boxplot(columns[i], ax=axes.flat[i], grid=False)

    upper_whisker = min(data[columns[i]].describe()['75%'] + 
                        1.5*(data[columns[i]].describe()['75%'] - 
                             data[columns[i]].describe()['25%']), data[columns[i]].max())
    lower_whisker = max(data[columns[i]].describe()['25%'] - 
                    1.5*(data[columns[i]].describe()['75%'] - 
                         data[columns[i]].describe()['25%']), data[columns[i]].min())
    axes.flat[i].text(0.25, 0.7, 
                      "Cтатистический\n минимум: {:.2f}\n Cтатистический\n максимум: {:.2f}".format(lower_whisker, upper_whisker), 
                      horizontalalignment='center', 
                      verticalalignment='center', 
                      transform=axes.flat[i].transAxes)

plt.show()
```

По графикам видно, что во всех параметрах, кроме высоты потолка, выбросы наблюдаются только выше статистического максимума. Для высоты потолка выбросы имеются также ниже статистического минимума.

#### тип этажа квартиры (значения — «‎первый», «последний», «другой»

```python
def floor_type(floor, floors_total):
    if floor==1:
        return 'первый'
    elif pd.isna(floors_total): 
        #у нас остались пропуски в столбце floors_total, поэтому мы точно не можем классифицировать этаж, если он не первый
        return np.nan
    elif floor==floors_total:
        return 'последний'
    else:
        return 'другой'
    
data['floor_type'] = data.apply(lambda x: floor_type(x['floor'], x['floors_total']), axis=1)
```

Населенные пункты с наибольшим числом объявлений - это Санкт-Петербург, посёлок Мурино, посёлок Шушары, Всеволожск, Пушкин, Колпино, посёлок Парголово, Гатчина, деревня Кудрово, Выборг.
